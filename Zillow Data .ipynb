```python
# Import all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Import CSV data from file
zillow_data_file = Path("zillow_values_2024.csv") 

zillow_data = pd.read_csv(zillow_data_file)

zillow_data.head(20)
```

```python
# Drop unwanted columns/rows
# Drop first row that lists United States, since we are looking for individual entries' values
zillow_data.drop([0], inplace=True)
```

```python
# Drop RegionID, SizeRank, RegionType as they are not used later
zillow_data.drop(columns=['RegionID','SizeRank','RegionType'], inplace=True)
zillow_data.head()
```

```python
# Drop any data before 2014
test_df = zillow_data.drop(zillow_data.columns[2:217],axis=1)
test_df.head()
```

```python
# split the RegionName value to get rid of state
test_df[['City', 'State']] = test_df['RegionName'].str.split(',', expand=True)
test_df.drop(columns=['RegionName','StateName'], inplace=True)
test_df.head()
```

```python
# check for null values
null_counts = test_df.isnull().sum()
# Print the count of null values in each column
print("Null counts in each column:")
for column, count in null_counts.items():
    print(f"{column}: {count}")
# in this case, there are few nulls so we will proceed without dropping more rows
```

```python
# Calculate the yearly average price for each city
# create list of years of interest
year_list=[2014,2015,2016,2017,2018,2019,2020,2021,2022,2023,2024]

# create for loop to move through list of years

for year in year_list:
    # use current year to create a term for regex search
    r_term=f'{year}.*'
    
    # filter columns to select only the ones with the desired year
    disposable_df=test_df.filter(regex= r_term)
    
    # create string to name a new column
    avg_term = f'{year}_avg'
    
    # add the average column to the original dataframe, skipping any missing values
    test_df[avg_term] = disposable_df.mean(axis=1, skipna=True)
    
test_df.head()

# average the columns in each group and append to the dataframe
```

```python
# drop the monthly data after the averages have been added, creating a clean dataframe
zillow_data_clean = test_df[['City'
                             ,'State'
                             ,'2014_avg'
                             ,'2015_avg'
                             ,'2016_avg'
                             ,'2017_avg'
                             ,'2018_avg'
                             ,'2019_avg'
                             ,'2020_avg'
                             ,'2021_avg'
                             ,'2022_avg'
                             ,'2023_avg'
                             ,'2024_avg']]

zillow_data_clean.head()
```

```python
columns_to_clean = ['State']
for column in columns_to_clean:
    zillow_data_clean[column] = zillow_data_clean[column].str.replace(" ", "")
zillow_data_clean
```

```python
zillow_data_clean.dropna(inplace=True)
zillow_data_clean
```

```python
columns_to_clean = ['2014_avg', '2015_avg', '2016_avg', '2017_avg', '2018_avg', '2019_avg', '2020_avg', '2021_avg', '2022_avg', '2023_avg', '2024_avg']

for column in columns_to_clean:
    zillow_data_clean[column] = zillow_data_clean[column].astype(int)

zillow_data_clean
```

```python
# Export to csv
zillow_data_clean.to_csv('housing.csv', index=False)
```
